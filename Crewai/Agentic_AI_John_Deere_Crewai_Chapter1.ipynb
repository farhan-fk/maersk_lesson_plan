```python
from crewai import Agent, Task, Crew
```

```python
import os

# Directly set your API key and model
os.environ["OPENAI_API_KEY"] = "your_openai_api_key_here"
os.environ["OPENAI_MODEL_NAME"] = "gpt-4o-mini"  # Use a valid model ID
```

```python
planner = Agent(
    role="Content Planner",
    goal="Plan engaging and factually accurate content on {topic}",
    backstory="You're working on planning a blog article "
              "about the topic: {topic}."
              "You collect information that helps the "
              "audience learn something "
              "and make informed decisions. "
              "Your work is the basis for "
              "the Content Writer to write an article on this topic.",
    allow_delegation=False,
	verbose=True
)
```

```python
writer = Agent(
    role="Content Writer",
    goal="Write insightful and factually accurate "
         "opinion piece about the topic: {topic}",
    backstory="You're working on a writing "
              "a new opinion piece about the topic: {topic}. "
              "You base your writing on the work of "
              "the Content Planner, who provides an outline "
              "and relevant context about the topic. "
              "You follow the main objectives and "
              "direction of the outline, "
              "as provide by the Content Planner. "
              "You also provide objective and impartial insights "
              "and back them up with information "
              "provide by the Content Planner. "
              "You acknowledge in your opinion piece "
              "when your statements are opinions "
              "as opposed to objective statements.",
    allow_delegation=False,
    verbose=True
)
```

```python
editor = Agent(
    role="Editor",
    goal="Edit a given blog post to align with "
         "the writing style of the organization. ",
    backstory="You are an editor who receives a blog post "
              "from the Content Writer. "
              "Your goal is to review the blog post "
              "to ensure that it follows journalistic best practices,"
              "provides balanced viewpoints "
              "when providing opinions or assertions, "
              "and also avoids major controversial topics "
              "or opinions when possible.",
    allow_delegation=False,
    verbose=True
)
```

```python
plan = Task(
    description=(
        "1. Prioritize the latest trends, key players, "
            "and noteworthy news on {topic}.\n"
        "2. Identify the target audience, considering "
            "their interests and pain points.\n"
        "3. Develop a detailed content outline including "
            "an introduction, key points, and a call to action.\n"
        "4. Include SEO keywords and relevant data or sources."
    ),
    expected_output="A comprehensive content plan document "
        "with an outline, audience analysis, "
        "SEO keywords, and resources.",
    agent=planner,
)
```

```python
write = Task(
    description=(
        "1. Use the content plan to craft a compelling "
            "blog post on {topic}.\n"
        "2. Incorporate SEO keywords naturally.\n"
		"3. Sections/Subtitles are properly named "
            "in an engaging manner.\n"
        "4. Ensure the post is structured with an "
            "engaging introduction, insightful body, "
            "and a summarizing conclusion.\n"
        "5. Proofread for grammatical errors and "
            "alignment with the brand's voice.\n"
    ),
    expected_output="A well-written blog post "
        "in markdown format, ready for publication, "
        "each section should have 2 or 3 paragraphs.",
    agent=writer,
)
```

```python
edit = Task(
    description=("Proofread the given blog post for "
                 "grammatical errors and "
                 "alignment with the brand's voice."),
    expected_output="A well-written blog post in markdown format, "
                    "ready for publication, "
                    "each section should have 2 or 3 paragraphs.",
    agent=editor
)
```

```python
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan, write, edit],
    verbose=2
)
```

```python
result = crew.kickoff(inputs={"topic": "Artificial Intelligence"})
```

```python
from IPython.display import Markdown
display(Markdown(result))
```

```python
#Smaple Excercise for Participants
# Used Case: Farmer Advisory Notes





# ---------------------------
# Agents
# ---------------------------
query_analyzer = Agent(
    role="Query Analyzer",
    goal=(
        "Understand a farmer's question like '{user_query}' and turn it into a concise brief: "
        "what they want, likely conditions in India, and the most probable causes or levers."
    ),
    backstory=(
        "You triage farmer questions for an agriculture advisory desk in India. "
        "You quickly clarify the core intent and extract the most relevant operational factors "
        "(tractor usage, soil/terrain, crop stage, irrigation pattern, maintenance habits). "
        "You are precise and practical. Do not reveal internal reasoning."
    ),
    allow_delegation=False,
    verbose=True
)

advisor = Agent(
    role="Agriculture Advisor",
    goal=(
        "Write a short, fact-based advisory to reduce fuel cost in tractor operations for Indian conditions."
    ),
    backstory=(
        "You provide actionable, brand-neutral recommendations for smallholders and contractors in India. "
        "You focus on low-cost, high-impact practices (e.g., correct gear/throttle combo, ballast & tire pressure, "
        "draft control, field route planning, implement matching, routine maintenance). "
        "Be practical and safety-conscious; avoid guarantees; don’t show internal chain-of-thought."
    ),
    allow_delegation=False,
    verbose=True
)

editor = Agent(
    role="Editor (Local-Friendly Tone)",
    goal="Simplify the advisory into two short paragraphs in clear, friendly Indian English.",
    backstory=(
        "You make content easy for farmers and dealer staff. Keep it simple and encouraging. "
        "Use plain English with occasional helpful Hindi terms in parentheses (e.g., 'pressure (dabav)'). "
        "Avoid jargon and brand claims. Keep to two paragraphs total."
    ),
    allow_delegation=False,
    verbose=True
)

# ---------------------------
# Tasks
# ---------------------------
analyze_query = Task(
    description=(
        "From the user query '{user_query}', produce a brief. Include:\n"
        "1) Problem summary (1–2 lines)\n"
        "2) Operating context assumptions for Indian farms (2–4 bullets)\n"
        "3) Likely fuel drivers (3–6 bullets; e.g., improper gear/throttle, over/under ballast, low tire pressure, "
        "mis-matched implement width, poor field routing, overdue maintenance, PTO mismatch)\n"
        "4) Key constraints/unknowns (bullets)\n"
        "Do not include sources or URLs. Do not expose internal reasoning."
    ),
    expected_output=(
        "A structured brief with headings: Problem Summary, Context Assumptions, Likely Fuel Drivers, Constraints/Unknowns."
    ),
    agent=query_analyzer,
)

draft_advice = Task(
    description=(
        "Using the analyzer brief, write a **short, fact-based advisory** focused on reducing tractor fuel cost. "
        "Requirements:\n"
        "- Indian conditions; low-cost actions first.\n"
        "- Include specific, safe, practical steps (e.g., select gear so engine runs near rated speed with moderate throttle; "
        "maintain correct tire pressure; optimize ballast; plan field passes; match implement width & PTO rpm; "
        "clean air filters; timely engine oil & filter changes; avoid idling).\n"
        "- No guarantees, no brand claims, no numbers unless widely accepted ranges.\n"
        "- Keep the advisory content to ~120–160 words before editing.\n"
        "Produce a clean draft (no headings)."
    ),
    expected_output="A concise advisory draft (~120–160 words), neutral and practical.",
    agent=advisor,
    context=[analyze_query]
)

edit_note = Task(
    description=(
        "Edit the advisory into **exactly two short paragraphs** (together ~120–160 words). "
        "Use clear Indian English, warm tone, minimal jargon. "
        "Optionally add 1–2 parenthetical Hindi terms where helpful. "
        "Remove any meta text, headings, or bullet lists. "
        "Final output must be just the two paragraphs."
    ),
    expected_output="Exactly two short paragraphs of final advisory text, no headings or bullets.",
    agent=editor,
    context=[draft_advice]
)

# ---------------------------
# Crew & Run
# ---------------------------
crew = Crew(
    agents=[query_analyzer, advisor, editor],
    tasks=[analyze_query, draft_advice, edit_note],
    verbose=2
)

# Example run
result = crew.kickoff(inputs={
    "user_query": "How to reduce fuel cost in tractor operations?"
})
print(result)
```

```python
# Case 3: --- CrewAI: Tractor Maintenance Tips Generator (LLM-only) -------------------
# If needed:
# !pip install crewai==0.28.8 crewai_tools==0.1.6 langchain_community==0.0.29



# ⛔️ Avoid hardcoding your key in code / repos
# import os
# os.environ["OPENAI_API_KEY"] = "YOUR_KEY"
# os.environ["OPENAI_MODEL_NAME"] = "gpt-4o-mini"  # or your chosen model

# ---------------------------
# Agents
# ---------------------------
planner = Agent(
    role="Maintenance Planner",
    goal=(
        "Design a practical, pre-{season} tractor maintenance checklist for {region} "
        "that prioritizes safety, reliability, and low-cost preventive actions."
    ),
    backstory=(
        "You are a service engineer who creates seasonal preventive maintenance checklists for farmers. "
        "You focus on the highest-impact checks for Indian conditions: engine oil & filters, air intake, "
        "fuel system water separation, battery & electrical protection, tire pressure & ballast, "
        "brakes, hydraulics, PTO & driveline, sheet metal rust protection, and storage practices. "
        "You keep steps simple, safe, and actionable."
    ),
    allow_delegation=False,
    verbose=True
)

writer = Agent(
    role="Technical Writer",
    goal=(
        "Expand the planner's outline into a clear Markdown article titled "
        "'Top 5 Tractor Maintenance Tips Before Monsoon Season', explaining what to do and why it helps."
    ),
    backstory=(
        "You turn engineering checklists into farmer-friendly guides. "
        "You explain the reasoning behind each step and offer practical, low-cost tips. "
        "You avoid brand claims and guarantees; you focus on safety and reliability."
    ),
    allow_delegation=False,
    verbose=True
)

editor = Agent(
    role="Editor (Farmer Communication)",
    goal="Polish the article into clear Indian English with simple, encouraging language.",
    backstory=(
        "You simplify technical material for Indian farmers and dealer staff. "
        "You keep headings, shorten long sentences, remove jargon, and ensure a warm, helpful tone. "
        "You may add an occasional Hindi term in parentheses for clarity (e.g., pressure (dabav))."
    ),
    allow_delegation=False,
    verbose=True
)

# ---------------------------
# Tasks
# ---------------------------
plan = Task(
    description=(
        "Create a 5-point pre-{season} tractor maintenance checklist for {region}. For each point:\n"
        "1) Give a short title (e.g., 'Check Engine Oil & Filters').\n"
        "2) Provide 1–2 lines on what to do (specific, safe, low-cost).\n"
        "3) Provide 1–2 lines on why it matters (reliability, fuel use, safety).\n"
        "Prioritize items relevant to rain/humidity (water ingress, corrosion, traction, electrics). "
        "Avoid brand claims, avoid guarantees, no external links."
    ),
    expected_output=(
        "A structured outline with exactly 5 checklist items. "
        "Each item has: Title, What to Do (1–2 lines), Why it Matters (1–2 lines)."
    ),
    agent=planner,
)

write = Task(
    description=(
        "Using the planner's outline, write a Markdown article titled:\n"
        "# Top 5 Tractor Maintenance Tips Before Monsoon Season\n\n"
        "Requirements:\n"
        "- Keep Indian context and {region} conditions in mind.\n"
        "- Include a short intro (3–5 sentences) and a brief closing (2–3 sentences).\n"
        "- For each of the 5 tips, add a '##' subheading matching the title in the outline, "
        "then write 2–3 short paragraphs: (a) what to do, (b) why it helps, (c) simple pro tip if useful.\n"
        "- Focus on safety (e.g., cool engine before checks, disconnect battery when needed, use PPE), practicality, and low-cost actions.\n"
        "- Avoid brand claims, pricing, and guarantees; do not invent specific model specs.\n"
        "- Target length ~600–800 words."
    ),
    expected_output="A well-structured Markdown article with title, intro, five '##' sections, and a closing.",
    agent=writer,
    context=[plan]
)

edit = Task(
    description=(
        "Edit the Markdown article to ensure clear Indian English, farmer-friendly tone, and readability:\n"
        "- Keep headings and structure; shorten long sentences; remove jargon.\n"
        "- Optionally add 1–2 helpful Hindi terms in parentheses where it improves clarity.\n"
        "- Ensure safety-first language and neutral, non-branded guidance.\n"
        "- Output must remain Markdown."
    ),
    expected_output="Publication-ready Markdown article, clean and easy to read.",
    agent=editor,
    context=[write]
)

# ---------------------------
# Crew & Sample Run
# ---------------------------
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan, write, edit],
    verbose=2
)

# Example run (you can change inputs as needed)
inputs = {
    "season": "Monsoon",
    "region": "India"
}

result = crew.kickoff(inputs=inputs)

# If you are in a notebook and want rich display:
try:
    from IPython.display import Markdown, display
    display(Markdown(result))
except Exception:
    print(result)
```
